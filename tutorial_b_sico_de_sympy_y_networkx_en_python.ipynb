# 1. Representación de matrices con SymPy

En SymPy las matrices se representan con la función `Matrix()`.

```python
A = sp.Matrix([
[1,2],
[3,4]
])

B = sp.Matrix([
[5,6],
[7,8]
])

A, B
```

# 7. Diagramas de transición con NetworkX

Un **diagrama de transición** representa los cambios posibles entre estados.
Cada estado es un nodo y cada transición es una arista dirigida.

```python
G = nx.DiGraph()
```

```python
# agregar nodos
G.add_nodes_from(["Estado 1","Estado 2","Estado 3"])
```

```python
# agregar transiciones
G.add_edge("Estado 1","Estado 2")
G.add_edge("Estado 2","Estado 3")
G.add_edge("Estado 3","Estado 1")
```

```python
pos = nx.spring_layout(G)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=2000,
    font_size=10,
    arrows=True
)

plt.show()
```
